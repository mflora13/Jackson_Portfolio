```python
import pandas as pd
import graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from graphviz import Source
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('course1_data_emails.csv')
```

```python
df = pd.get_dummies(df, drop_first=True)
train_cols = df.drop('clicked',axis=1)
```

```python
tree_clf = DecisionTreeClassifier(max_depth=4,
                                 class_weight='balanced',
                                 min_impurity_decrease=0.001)
```

```python
tree_clf.fit(train_cols, df['clicked'])
```

```python
plt.figure(figsize=(15,8))
tree.plot_tree(tree_clf);
#hard to see
```

```python
export_graphviz(tree_clf, out_file='tree.dot',feature_names=train_cols.columns,
               proportion=True,rotate=True)

s=Source.from_file('tree.dot')
s.view()
```
